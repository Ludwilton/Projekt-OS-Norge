```python
# E)

medelvärde = athlete_events["Age"].mean()

median = athlete_events["Age"].median()

min = athlete_events["Age"].min()

max = athlete_events["Age"].max()

standardavikelse = athlete_events["Age"].std()

print (f"{medelvärde:.2f}")
print (median)
print (min)
print (max)
print (f"{standardavikelse:.2f}")
```

```python


medalj_data = athlete_events.dropna(subset=["Medal"])


medaljer_per_land = medalj_data.groupby("NOC")["Medal"].count()


topp_10_lander = medaljer_per_land.nlargest(10)


plt.figure(figsize=(10, 6))
topp_10_lander.plot(kind="bar", color="skyblue")
plt.suptitle("Topp 10 Länder med Flest Medaljer")
plt.xlabel("Land")
plt.ylabel("Antal Medaljer")
plt.xticks(rotation=45)
plt.show()

```

```python
sex_value_count= athlete_events["Sex"].value_counts()

plt.figure(figsize=(8, 8))
sex_value_count.plot(kind="pie", autopct="%1.1f%%", startangle=50, colors=["skyblue", "pink"])
plt.title =("Könsfördelning")
plt.show
```
